```python
x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_scaled = pd.DataFrame(x_scaled)
df_scaled
```

```python
# Feature engineering function to define time series data
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
# Small Lstms for forcasting the features in the target timeline.
step_size = 3
y_predict = []
n_features = 1

for indx in df_scaled.columns:
    X, y = split_sequence(df_scaled[indx], step_size)
#     print (X.shape)
    X_train = X[0:X.shape[0], :]
    y_train = y[0:X.shape[0]]
    X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], n_features))
    # define model
    model = Sequential()
    model.add(Bidirectional(LSTM(50, activation='relu'), input_shape=(step_size, n_features)))
#     model.add(LSTM(50, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    # fit model
    model.fit(X_train, y_train, epochs=200, verbose=0)
    # demonstrate prediction
    x_input = X_train[-1]
    x_input = x_input.reshape((1, step_size, n_features))
    yhat = model.predict(x_input, verbose=0)
    y_predict.append(yhat)
    print (x_input)
    print (yhat)
print (y_predict)
# 0.542373	0.148823	0.036053	0.012171	0.002585	0.176960	0.015689	0.005756	0.149334	0.112283
```
